```python
import pandas as pd
```

```python
data=pd.read_csv('/content/adult_data_mini.csv')
```

```python
sid = 2443834
n = sid % 10
if n == 0:
  n = 10
print(f"The value of n is: {n}")
```

```python
relationship_hours_group = data.groupby(['relationship', 'hours-per-week']).size()
display(relationship_hours_group)
```

```python
def reduce_hours(hours):
  return hours - n

data['hours-per-week_reduced'] = data['hours-per-week'].apply(reduce_hours)
display(data[['hours-per-week', 'hours-per-week_reduced']].head())
```

```python
relationship_reduced_hours_group = data.groupby(['relationship', 'hours-per-week_reduced']).size()
display(relationship_reduced_hours_group)
```
